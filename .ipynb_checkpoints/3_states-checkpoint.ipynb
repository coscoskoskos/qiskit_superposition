```python
import numpy as np
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit import IBMQ, execute, Aer
from qiskit.tools.visualization import plot_histogram
import math
```

```python
#3 qubits superposition
q = QuantumRegister(2, 'q')  #00, 01, 11
c = ClassicalRegister(2, 'c')
qc = QuantumCircuit(q, c)

theta = 2 * np.arccos(1 / np.sqrt(3))

qc.ry(theta, q[0])
qc.ch(q[0], q[1])

qc.measure(q[0:2], c[0:2])

r = execute(qc, Aer.get_backend('qasm_simulator')).result()
rc = r.get_counts()
print(rc)
plot_histogram(rc)
```

```python
q = QuantumRegister(2, 'q')  #01, 10, 11
c = ClassicalRegister(2, 'c')
qc = QuantumCircuit(q, c)

theta = 2 * np.arccos(1 / np.sqrt(3))

qc.ry(theta, q[0])
qc.ch(q[0], q[1])
qc.x(q[1])

qc.measure(q[0:2], c[0:2])

r = execute(qc, Aer.get_backend('qasm_simulator')).result()
rc = r.get_counts()
print(rc)
plot_histogram(rc)
```
